```python
import numpy as np 
import pandas as pd
import sys
from astropy.io import fits
import plotly.graph_objects as go
sys.path.append('../../dash_gaia_orbits/')

# Package imports
from dash_gaia_orbits.dash_test import create_dash_app
from dash_gaia_orbits import StarClusterData, StarClusterCollection, StarClusters3DPlotter
```

```python
df = pd.read_csv('/Users/cam/Downloads/cluster_sample_data.csv')
df = df[['name', 'family', 'age_myr', 'U', 'V', 'W', 'x_helio', 'y_helio', 'z_helio', 'n_stars']]
df = df.rename(columns={'x_helio' : 'x', 'y_helio' : 'y', 'z_helio' : 'z'})

alphaPer_family_clusters = df.loc[df['family'] == 'alphaPer']
carina = alphaPer_family_clusters.loc[alphaPer_family_clusters['name'] == 'CAR']
alphaPer_similar_age_car = alphaPer_family_clusters.loc[(alphaPer_family_clusters['age_myr'].between(40, 50)) & (alphaPer_family_clusters['name'] != 'CAR')] # just a guess for what age range is similar to CAR (45 Myr)
```

```python
sun = pd.DataFrame({'name' : ['Sun'], 'family' : ['Sun'], 'age_myr' : [8000], 'U' : [0.], 'V' : [0.], 'W' : [0.], 'x' : [0.], 'y' : [0.], 'z' : [0.], 'n_stars' : [1]})

sun_data = StarClusterData(sun, data_name = 'Sun', min_size = 0, max_size = 10, color = 'yellow', opacity = 1., marker_style = 'circle')
carina_data = StarClusterData(carina, data_name = 'Carina', min_size = 0, max_size = 15, color = 'violet', opacity = 1., marker_style = 'cross')
similar_age_carina_data = StarClusterData(alphaPer_similar_age_car, data_name = 'Similar Age to Carina', min_size = 0, max_size = 12, color = 'black', opacity = 1., marker_style = 'circle')
alphaPer_fam_data = StarClusterData(alphaPer_family_clusters, data_name = 'Alpha Per Family', min_size = 0, max_size = 5, color = 'black', opacity = 1., marker_style = 'circle')

sc = StarClusterCollection([sun_data, alphaPer_fam_data, carina_data, similar_age_carina_data])
plot_3d = StarClusters3DPlotter(data_collection = sc, xyz_widths = (300, 300, 300), figure_theme = 'light')

save_name = '/Users/cam/Downloads/carina_focused_movie.html'
time_int = np.round(np.arange(0., -60.5, -0.5), 1)
fig3d_m6 = plot_3d.generate_3d_plot(
    time = time_int,
    show = True, 
    save_name =save_name, 
    static_traces = None, 
    static_traces_times = None, 
    focus_group = 'Carina',
    fade_in_time = 5 # Myr
)
```
